```python
import pandas as pd
import numpy as np
```

```python
arr = []
for i in range(10):
    data = np.random.random(size=5)
    arr.append(data)
df = pd.DataFrame(arr)
```

```python
df['mean'] = 0.0
for i in range(df.shape[0]):
    summ = 0.0
    cnt = 0
    for j in range(df.shape[1] - 1):
        if df.iloc[i, j] > 0.3:
            summ += df.iloc[i, j]
            cnt += 1
    if cnt != 0:
        df.iloc[i,5] = summ / cnt
    
df
```

```python
df = pd.DataFrame(np.random.random((10, 5)))
mask = df > 0.3
df['mean'] = df.where(mask).mean(axis=1).fillna(0.0)
df
```
